```python
import pandas as pd
import numpy as np
```

```python
suppliers=pd.read_csv("./data/northwind-mongo-master/suppliers.csv",usecols=["SupplierID","CompanyName","City"])
```

```python
products=pd.read_csv("./data/northwind-mongo-master/products_del_2.csv",usecols=["ProductID","ProductName","SupplierID"])
```

```python
suppliers.merge(products,on="SupplierID")
```

```python
suppliers.set_index("SupplierID",inplace=True)
```

```python
suppliers.merge(products, right_on="SupplierID",left_index=True)
```

```python
products.set_index("SupplierID",inplace=True)
```

```python
suppliers.merge(products, right_index=True, left_index=True)
```

```python
suppliers.reset_index(inplace=True)
```

```python
suppliers.merge(products, left_on="SupplierID",right_index=True)
```

```python
products.reset_index(inplace=True)
```

```python
products.rename({"SupplierID":"supplierid"},axis="columns",inplace=True)
```

```python
suppliers.merge(products, left_on="SupplierID",right_on="supplierid")
```
