```python
import pandas as pd
import numpy as np
```

```python
products=pd.read_csv("./data/WA_Sales_Products_2012-14.csv")
products.head()
```

```python
groups=products.groupby(by=["Retailer country","Year"])
```

```python
groups["Revenue"].sum().head()
```

```python
groups["Quantity"].sum().head()
```

```python
groups["Gross margin"].sum().head()
```

```python
groups.Revenue.sum().head()
```

```python
groups.Quantity.sum().head()
```

```python
groups.agg({"Revenue" : "sum"}).head()
```

```python
groups.agg({"Gross margin" : "mean"}).head()
```

```python
groups.agg({"Gross margin" : "mean",
           "Revenue" : "sum",
           "Quantity" : "sum"}).head()
```

```python
groups.agg({"Gross margin" : ["sum","min","max"],
           "Revenue" : ["sum","min","max"],
           "Quantity" : "sum"}).head()
```

```python
my_aggregation=["sum","min","max"]
```

```python
groups.agg({"Gross margin" : my_aggregation,
           "Revenue" : my_aggregation,
           "Quantity" : "sum"}).head()
```
