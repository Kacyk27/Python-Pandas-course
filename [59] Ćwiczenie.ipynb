```python
import pandas as pd
import numpy as np
```

```python
suppliers=pd.read_csv("./data/northwind-mongo-master/suppliers.csv",usecols=["SupplierID","CompanyName","City"])
```

```python
products=pd.read_csv("./data/northwind-mongo-master/products_del_2.csv",usecols=["ProductID","ProductName","SupplierID"])
```

```python
supprod=suppliers.merge(products, on="SupplierID",indicator=True)
supprod
```

```python
print(len(suppliers))
print(len(products))
print(len(supprod))
```

```python
supProdLeft = suppliers.merge(products, how="left", on="SupplierID",indicator=True)
```
