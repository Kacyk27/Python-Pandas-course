```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
bar = pd.read_csv("./data/weather_barcelona.csv",index_col="Date")
```

```python
rom = pd.read_csv("./data/weather_rome.csv",index_col="Date")
```

```python
ams = pd.read_csv("./data/weather_amsterdam.csv",index_col="Date")
```

```python
bar.head()
```

```python
tempMax=pd.DataFrame(index=bar.index)
```

```python
tempMax["Barcelona"]=bar["TempMax"]
tempMax["Rome"]=rom["TempMax"]
tempMax["Amstredam"]=ams["TempMax"]
```

```python
tempMax.head()
```

```python
tempMax.plot()
```

```python
tempMax.plot(figsize=(15,5))
```

```python
tempMax.plot(figsize=(15,5), subplots=True)
```

```python
tempMax.plot(figsize=(15,5), use_index=False)
```

```python
tempMax.plot(figsize=(15,5), title="Max temperature by city")
```

```python
tempMax.plot(figsize=(15,5), legend=False)
```

```python
tempMax.plot(figsize=(15,5), logy=True)
```
