```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
marathon = pd.read_csv("./data/marathon_results_2017.csv", usecols=["Age","M/F","Country","40K"])
marathon["TimeSeconds"] = marathon["40K"].apply(lambda x: pd.Timedelta(x).total_seconds())
marathon.head(5)
    
```

```python
marathon.plot.scatter(x="Age",y="TimeSeconds")
```

```python
marathonM= marathon[marathon["M/F"]=="M"]
marathonM.head()
```

```python
marathonF= marathon[marathon["M/F"]=="F"]
marathonF.head()
```

```python
ax= marathonM.plot.scatter(x="Age",y="TimeSeconds",color="Orange",label="Male")
marathonF.plot.scatter(x="Age",y="TimeSeconds",color="Blue",label="Female",ax=ax)
```

```python
marathonM.plot.hexbin(x="Age",y="TimeSeconds",gridsize=25)
```

```python
marathonF.plot.hexbin(x="Age",y="TimeSeconds",gridsize=25)
```
